```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('retail_sales_dataset.csv')

# First look at data
print("First 5 rows:")
print(df.head())

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns.tolist())
```

```python
print("Basic Statistics:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Step 3: Sales Visualization
import matplotlib.pyplot as plt

# Chart 1: Sales by Product Category
product_sales = df.groupby('Product Category')['Total Amount'].sum()

plt.figure(figsize=(8,5))
product_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Sales by Category:")
print(product_sales)
```

```python
# Chart 2: Monthly Sales Trend
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

monthly_sales = df.groupby('Month')['Total Amount'].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(kind='line',
                   marker='o',
                   color='green')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Chart 3: Sales by Gender
gender_sales = df.groupby('Gender')['Total Amount'].sum()

plt.figure(figsize=(6,6))
gender_sales.plot(kind='pie',
                  autopct='%1.1f%%',
                  colors=['pink','skyblue'])
plt.title('Sales by Gender')
plt.ylabel('')
plt.tight_layout()
plt.show()

print("Gender Sales:")
print(gender_sales)
```

```python
# Final: Recommendations
print("=" * 40)
print("EDA INSIGHTS & RECOMMENDATIONS")
print("=" * 40)

print("\n1. TOP SELLING CATEGORY:")
print("   Electronics = ₹1,56,905")

print("\n2. BEST SALES MONTH:")
best_month = monthly_sales.idxmax()
print(f"   Month {best_month}")

print("\n3. CUSTOMER INSIGHTS:")
print("   Female customers = 51.1%")
print("   Male customers = 48.9%")

print("\n4. RECOMMENDATIONS:")
print("   - Focus on Electronics")
print("   - Target female customers")
print("   - Boost sales in weak months")
print("=" * 40)
```
